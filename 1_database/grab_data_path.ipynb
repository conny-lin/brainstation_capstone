## get image file paths

```python
# get image files paths
# set up dataframe
MWTDB['png_path'] = ['NaN']*MWTDB.shape[0]
MWTDB['shanespark_path'] = ['NaN']*MWTDB.shape[0]
MWTDB['drunkposture_path'] = ['NaN']*MWTDB.shape[0]
MWTDB['drunkposture2_path'] = ['NaN']*MWTDB.shape[0]
MWTDB['mwtfolder_exist'] = [True]*MWTDB.shape[0]
MWTDB
```

```python
# for every mwt plate, find file path and put it in the database
report_count=0
# for i in range(,100):
for i in range(0,MWTDB.shape[0]):
    pp = MWTDB.loc[i,'mwtpath'] # get path to mwt plate file
    platename = MWTDB.loc[i,'mwtname'] # get plate name
    
    # report progress every 50 plates
    if report_count%100 == 0:
        print(f'looking up {i}')
        
    if os.path.isdir(pp):
        # get png file
        dat_file_path = [_ for _ in os.listdir(pp) if _.endswith(r'.png')]
        if dat_file_path:
            MWTDB.loc[i,'png_path'] = os.path.join(pp,dat_file_path[0])

        # get shanespark.dat file
        dat_file_path = [_ for _ in os.listdir(pp) if _.endswith(r'.shanespark.dat')]
        if dat_file_path:
            MWTDB.loc[i,'shanespark_path'] = os.path.join(pp,dat_file_path[0])
            
        # get drunkposture.dat
        dat_file_path = [_ for _ in os.listdir(pp) if _.endswith(r'.drunkposture.dat')]
        if dat_file_path:
            MWTDB.loc[i,'drunkposture_path'] = os.path.join(pp,dat_file_path[0])
        
        # get drunkposture2.dat
        dat_file_path = [_ for _ in os.listdir(pp) if _.endswith(r'.drunkposture2.dat')]
        if dat_file_path:
            MWTDB.loc[i,'drunkposture2_path'] = os.path.join(pp,dat_file_path[0])
    else:
        MWTDB.loc[i,'mwtfolder_exist'] = False
        print(f'\tno folder')
    # get report count
    report_count +=1

display(MWTDB)
```

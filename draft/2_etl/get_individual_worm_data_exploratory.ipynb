````python
# get row numbers for all of the files
# declare empty array to store the row numbers
wormid_array = np.empty((fileDB_target.shape[0],1),dtype='int')
# loop through each files 
for i, p in enumerate(fileDB_target['path'], 0):
    # report process every 100 files
    if (i%100==0):
        print(f'processing #{i} files', end='\r')
    # create custom bash command
    bashCommand = "cat "+p+" | wc -l"
    # call bash
    output = subprocess.check_output(bashCommand, shell=True)
    # take out digits from the byte output
    wormid_array[i] = np.array(output,dtype='int')
wormid_array
````
